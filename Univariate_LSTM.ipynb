```python
# univariate lstm
from numpy import array
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/Temperature.csv")
```

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
 X, y = list(), list()
 for i in range(len(sequence)):
 # find the end of this pattern
  end_ix = i + n_steps
  # check if we are beyond the sequence
  if end_ix > len(sequence)-1:
    break
 # gather input and output parts of the pattern
  seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
  X.append(seq_x)
  y.append(seq_y)
 return array(X), array(y)
```

```python
# define input sequence
#raw_seq = [10, 20, 30, 40, 50, 60, 70, 80, 90]
# choose a number of time steps
n_steps = 3
# split into samples
X, y = split_sequence(df['MaxTemp'], n_steps)
```

```python
# summarize the data
for i in range(len(X)):
 print(X[i], y[i])
```

```python
# reshape from [samples, timesteps] into [samples, timesteps, features]
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))
```

```python
# define model

#This initializes a Keras Sequential model, which is a linear stack of layers
#Adds an LSTM layer with 50 units (neurons)
#activation='relu' ReLU (Rectified Linear Unit) is a common activation function that helps introduce non-linearity into the model.
#input_shape=(n_steps, n_features): Specifies the shape of the input data

model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(n_steps, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
# fit model
model.fit(X, y, epochs=200, verbose=0)
```

```python
# demonstrate prediction
x_input = array([28.5,27.4,20.6])
x_input = x_input.reshape((1, n_steps, n_features))
yhat = model.predict(x_input, verbose=0)
print(yhat)
```
